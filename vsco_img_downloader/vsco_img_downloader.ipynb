```python
import time
import threading
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import requests # to get image from the web
import shutil # to save it locally
```

```python
def download_image(image_url):
    filename = image_url.split("/")[-1]

    # Open the url image, set stream to True, this will return the stream content.
    r = requests.get(image_url, stream = True)

    # Check if the image was retrieved successfully
    if r.status_code == 200:
        # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
        r.raw.decode_content = True

        # Open a local file with wb ( write binary ) permission.
        with open(filename,'wb') as f:
            shutil.copyfileobj(r.raw, f)

        print('Image sucessfully Downloaded: ',filename)
    else:
        print('Image Couldn\'t be retreived')
```

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
f = open('sources.txt', "r")
txt = f.read()
items = txt.split('\n')
item_list_chunks = list(chunks(items, 50))
```

```python
def download_list(items):
    browser = webdriver.Firefox()
    for item in items:
        browser.get(item)
        image_url = WebDriverWait(browser,5).until(
            EC.presence_of_element_located((By.XPATH, '//img'))).get_attribute('src')
        print(image_url)
        if '.jpg' in image_url:
            download_image(image_url)
    browser.close()
```

```python
for litem_list_chunk in item_list_chunks:
    t = threading.Thread(target=download_list, args=[litem_list_chunk])
    t.start()
```
